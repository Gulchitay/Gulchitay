```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a Discrete-Time Impulse Response
n = np.arange(0, 20)
h_n = 0.5 ** n

# Step 2: Define the Step Input Signal
u_n = np.ones_like(n)

# Step 3: Convolution to Find the Step Response
step_response = np.cumsum(h_n)

# Step 4: Perform Analysis and Visualization
plt.figure(figsize=(10, 6))

# Part 1: Impulse Response
plt.subplot(3, 1, 1)
plt.stem(n, h_n, )
plt.title('Impulse Response h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')

# Part 2: Step Response
plt.subplot(3, 1, 2)
plt.stem(n, step_response, basefmt="", markerfmt="bo", linefmt="y--", label="Step redsponse" )
plt.title('Step Response')
plt.xlabel('n')
plt.ylabel('Step Response')

# Part 3: Validation via Convolution
conv_step_response = np.convolve(h_n, u_n)[:len(n)]
plt.subplot(3, 1, 3)
plt.stem(n, conv_step_response, basefmt="", markerfmt="bo", linefmt="r--", label="Convolution step redsponse" )
plt.title('Convolution Step Response')
plt.xlabel('n')
plt.ylabel('Convolved Response')

plt.tight_layout()
plt.show()

# Objective: Signal Quantization

# Part 1: Generate a Continuous Signal
A = 1  # Amplitude
f = 5  # Frequency in Hz
t = np.linspace(0, 1, 1000)  # Time vector
continuous_signal = A * np.sin(2 * np.pi * f * t)

# Plot the continuous signal
plt.figure(figsize=(10, 4))
plt.plot(t, continuous_signal, label='Continuous Signal', color='blue')
plt.title('Continuous Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.show()

# Part 2: Quantization
def quantize_signal(signal, bits):
    levels = 2 ** bits
    quantized_signal = np.round((signal + 1) * (levels - 1) / 2).clip(0, levels - 1)
    return quantized_signal * 2 / (levels - 1) - 1  # Scale back to original range

# Quantized signals
quantized_2bit = quantize_signal(continuous_signal, 2)
quantized_3bit = quantize_signal(continuous_signal, 3)
quantized_4bit = quantize_signal(continuous_signal, 4)

# Plot the original and quantized signals
plt.figure(figsize=(10, 6))
plt.plot(t, continuous_signal, label='Original Signal', color='blue', alpha=0.5)
#plt.step(t, quantized_2bit, label='2-bit Quantization', where='post', color='red', alpha=0.7)
#plt.step(t, quantized_3bit, label='3-bit Quantization', where='post', color='green', alpha=0.7)
plt.step(t, quantized_4bit, label='4-bit Quantization', where='post', color='orange', alpha=0.7)
plt.title('Quantization of Continuous Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()

# Part 3: Quantization Error
def quantization_error(original_signal, quantized_signal):
    return original_signal - quantized_signal

error_2bit = quantization_error(continuous_signal, quantized_2bit)

# Plot the quantization error for 2-bit quantization
plt.figure(figsize=(10, 4))
plt.plot(t, error_2bit, label='Quantization Error (2-bit)', color='purple')
plt.title('Quantization Error for 2-bit Quantization')
plt.xlabel('Time (s)')
plt.ylabel('Error')
plt.grid()
plt.legend()
plt.show()

error_3bit = quantization_error(continuous_signal, quantized_3bit)

# Plot the quantization error for 3-bit quantization
plt.figure(figsize=(10, 4))
plt.plot(t, error_3bit, label='Quantization Error (3-bit)', color='purple')
plt.title('Quantization Error for 3-bit Quantization')
plt.xlabel('Time (s)')
plt.ylabel('Error')
plt.grid()
plt.legend()
plt.show()

error_4bit = quantization_error(continuous_signal, quantized_4bit)

# Plot the quantization error for 4-bit quantization
plt.figure(figsize=(10, 4))
plt.plot(t, error_4bit, label='Quantization Error (4-bit)', color='purple')
plt.title('Quantization Error for 4-bit Quantization')
plt.xlabel('Time (s)')
plt.ylabel('Error')
plt.grid()
plt.legend()
plt.show()
```
